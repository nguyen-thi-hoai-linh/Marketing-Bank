```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hongnhungyknb1925/data_driven/main/bank-additional-full.csv", sep =";")
df.shape
```

```python
df.describe()
```

```python
df.columns
```

### Drop và số hóa các feature

```python
df = df.drop(["month","day_of_week","contact"],axis=1)
df["education"]=df["education"].replace(['basic.4y','high.school','basic.6y','basic.9y','professional.course','university.degree','illiterate',"unknown"],[0,1,2,3,4,5,6,np.nan])
```

```python
df.housing[df['housing']=='no']=0
df.housing[df['housing']=='yes']=1
df.housing[df['housing']=='unknown']=np.nan
```

```python
df.loan[df['loan']=='no']=0
df.loan[df['loan']=='yes']=1
df.loan[df['loan']=='unknown']=np.nan
```

```python
df["loan"]=df["loan"]+df["housing"]
df = df.drop("housing",axis=1)
```

```python
df.head()
```

```python
df.head()
```

Outlier

```python
Q1 = df['duration'].quantile(.25)
Q3 = df['duration'].quantile(.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[df['duration'] >= lower] 
df = df[df['duration'] <=upper] 
```

```python
Q1 = df['age'].quantile(.20)
Q3 = df['age'].quantile(.80)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[df['age'] >= lower] 
df = df[df['age'] <=upper] 
```

```python
Q1 = df['previous'].quantile(.20)
Q3 = df['previous'].quantile(.80)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[df['previous'] >= lower] 
df = df[df['previous'] <=upper]
```

```python
df["y"]=df["y"].replace("no",0)
df["y"]=df["y"].replace("yes",1)
```

```python
from pandas.core.algorithms import value_counts
df["y"].value_counts()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
trace0 = go.Box(y=df['duration'],name='Duration')
trace1 = go.Box(y=df['age'],name='Age')
trace2 = go.Box(y=df['previous'],name='Previous')

fig = make_subplots(rows=2, cols=3)
fig.append_trace(trace0, row = 1, col = 1)
fig.append_trace(trace1, row = 1, col = 2)
fig.append_trace(trace2, row = 1, col = 3)

fig.update_layout(width=800, height=400, title='Box Plot to check for outliers')

fig.show()
```

```python
y = df["y"]
X = df.drop(['y'], axis=1)
```

```python
o =['job','marital','default','poutcome','campaign']
labelencoder = LabelEncoder()
for c in o:
    X[c]      = labelencoder.fit_transform(X[c]) 
X.head()
```

```python
df.info()
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=9)
X=pd.DataFrame(imputer.fit_transform(X),columns=X.columns)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression() 
logmodel.fit(X_train,y_train)
logpred = logmodel.predict(X_test)
```

```python
accuracy_log = accuracy_score(y_test, logpred)
accuracy_log
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error

print('Mean Absolute Error      : ', metrics.mean_absolute_error(y_test, logpred))
print('Mean Squared  Error      : ', metrics.mean_squared_error(y_test, logpred))
print('Root Mean Squared  Error : ', np.sqrt(metrics.mean_squared_error(y_test, logpred)))
print('R Squared Error          : ', metrics.r2_score(y_test, logpred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion='gini') #criterion = entopy, gini
dtree.fit(X_train, y_train)
dtreepred = dtree.predict(X_test)
```

```python
accuracy_dtree = accuracy_score(y_test, dtreepred)
accuracy_dtree
```
